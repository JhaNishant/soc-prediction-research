```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import IterativeImputer

# Create an imputer
imputer = IterativeImputer(estimator=RandomForestRegressor(n_estimators=10, random_state=0))

# List of column names
cols = df.columns

# Create a copy of the dataframe
df_imputed = df.copy()

# Impute each column one by one
for col in cols:
    if df[col].isnull().sum() > 0:
        print(f"Imputing {col}...")
        df_imputed[col] = imputer.fit_transform(df[[col]])
        print(f"Finished imputing {col}.")

print("Imputation complete.")

# The result is a numpy array, so convert it back to a DataFrame
df_imputed = pd.DataFrame(df_imputed, columns=df.columns)
```
